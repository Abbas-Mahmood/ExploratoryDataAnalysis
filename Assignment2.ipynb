### Question 2 (3 marks)
The CSV data files "north.csv" and "south.csv" contain the time series of the average sea level observed in each month of the year in the northern and in the southern emisphere, respectively. The rows of the file are in the format:

        data,value

where ```data``` is a year and month in the format ```YYYYMM```, while ```value``` is a real number. 


#### Q2 (a)

Load the two data files "north.csv" and "south.csv" in two separate (and appropriate) variables. Plot the two time series using a figure with two distinct subplots.

**(1.5 marks)**

````python
### Q2 (a)
North = np.loadtxt('north.csv',delimiter=',')
South = np.loadtxt('south.csv',delimiter=',')

# assigning each column of data files to a separate variable. First column: Data, Second column: Values
N_Data = North[:,0]
N_Value = North[:,1]
S_Data = South[:,0]
S_Value = South[:,1]
# generating a figure with 2 sets of axes across 1 row and 2 columns.
fig,ax = plt.subplots(1,2)
# plotting the given time series on each set of axes
ax[0].plot(N_Data,N_Value,'-r');
ax[1].plot(S_Data,S_Value,'-b');

ax[0].set_title('North'); 
ax[1].set_title('South');
fig.autofmt_xdate()
````

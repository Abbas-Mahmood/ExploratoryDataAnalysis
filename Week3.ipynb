 ## Simple plotting


```python
# Use the numpy function loadtxt to load the file into a variable called StockPrice
StockPrice = np.loadtxt('Apple_Stock.csv')
print(StockPrice)
```

```python
# Now divide StockPrice into two variables X and Y
X = StockPrice[:,0]
Y = StockPrice[:,1]
```

```python
# create the first plot of X versus Y
plt.plot(X,Y); #dataset is scrambled which is causing plot issue.
plt.title('Apple Stock Price')
plt.xlabel('index')
plt.ylabel('Price')
```

```python
# create a scatter plot of X versus Y
plt.scatter(X,Y);
```

Clearly soemthing is not right with the line plot, we need to sort the first column. 

```python
# use the np.sort function to sort X into vartiable X_sorted
# and plot again
X_sorted = np.sort(X)
plt.plot(X_sorted,Y)
```

```python
# Now we will get it right, we need to sort Y the same way as we sorted X using argsort()
order = np.argsort(X) # returns indices that would sort array
X_sorted = X[order]
Y_sorted = Y[order]
print(order)
plt.plot(X_sorted,Y_sorted)
```
